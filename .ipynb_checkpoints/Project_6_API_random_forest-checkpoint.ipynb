## Part 5: Build the Decision Tree

#### 1. What is our target attribute? 

Our target attribute is rating

#### 2. Prepare the data and define the training set

```python
DF = pd.merge(left = top_50, right = df_m, left_on = 'tconst', right_on = 'index')
DF
```

```python
X = DF.ix[:,6:]
print X.shape
X.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cvec = CountVectorizer()
cvec.fit(X)
X = pd.DataFrame(cvec.transform(X).todense())
```

#### 2. Train the Model

```python
y = DF.rating.values.astype(int)
print len(y)
print y[:5]
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(class_weight='balanced')
tree.fit(X_train, y_train)
```

#### 3. Set up test data and test the model

```python
print tree.score(X_test, y_test)
```

#### 5. Check the results

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.cross_validation import cross_val_score
```

```python
predictions = tree.predict(X_test)
accuracy_score(y_test, predictions)
```

```python
scores = cross_val_score(tree, X, y, cv=3)
print 'All Scores', scores
print 'Average Score', np.mean(scores)
```

#### 6. What is overfitting and how are we at risk? 

```python
params = {'max_depth':[3,5,10,20], 'max_features':[1,2,3,'auto'], 
              'max_leaf_nodes':[5,6,7,8,9,10], 'min_samples_leaf':[1,3,5,7,10],
              'min_samples_split':[2,5,7]}
scoring = 'mean_squared_error'
```

```python
from sklearn.grid_search import GridSearchCV
model = DecisionTreeClassifier()
clf = GridSearchCV(model, params, scoring=scoring, verbose=True, n_jobs=-1)
clf.fit(X, y)
```
