# Part 2: Wrangle the text data

#### 1. Convert the listing identification numbers (tconst) from the first dataframe to a list

```python
idnumbers = top_50.tconst.values
idnumbers
```

#### 2. Scrape the reviews for the top 25 movies

*Hint*: Use a loop to scrape each page at once

```python
reviews = {}
for idn in idnumbers:
    reviews[idn] = imdb.get_title_reviews(idn, max_results=50)
```

```python
#reviews
reviews.items()
```

```python
reviews_df = pd.DataFrame(reviews).transpose()
reviews_df.head()
```

```python
reviews_df.columns = ['review_1','review_2','review_3','review_4','review_5','review_6','review_7','review_8','review_9',
                'review_10','review_11','review_12','review_13','review_14','review_15','review_16','review_17',
                'review_18','review_19','review_20','review_21','review_22','review_23','review_24','review_25', 'review_26','review_27','review_28','review_29','review_30','review_31','review_32','review_33','review_34',
                'review_35','review_36','review_37','review_38','review_39','review_40','review_41','review_42',
                'review_43','review_44','review_45','review_46','review_47','review_48','review_49','review_50']
reviews_df.head()
```

```python
reviews_df.info()
```

```python
reviews_df.review_1[1]
```

```python
for movie in range(1,len(reviews_df.columns)+1):
    review_clean = []
    for i in eval('reviews_df.review_'+str(movie)):
        review_clean.append(i.text)
    reviews_df['review_'+str(movie)] = review_clean
reviews_df.head()
```

```python
reviews_df.ix[0,0]
```

```python
for movie in range(1,len(reviews_df.columns)+1):
    review_clean = []
    for i in eval('reviews_df.review_'+str(movie)):
        review_clean.append(re.sub(r"[^a-zA-Z0-9 ]", "", i))
    reviews_df['review_'+str(movie)] = review_clean
reviews_df.head()
```
